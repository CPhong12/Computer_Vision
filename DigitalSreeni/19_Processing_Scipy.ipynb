```python
from skimage import io, img_as_ubyte
from scipy import ndimage
import numpy as np
from matplotlib import pyplot as plt
img =img_as_ubyte(io.imread('C:\Code\OpenCV\DigitalSreeni\Wow.jpg',as_gray=True))
```

```python
print(img.shape, img.dtype)
print(img[0:10,20:25])
```

```python
mean_gray = img.mean()
max_value = img.max()
min_value = img.min()

print('Min, Max and Mean are:', min_value, max_value, mean_gray)
```

### Flip, Rotate

```python
flippedLR = np.fliplr(img)
flippedUD = np.flipud(img)
```

```python
plt.subplot(2,1,1)
plt.imshow(img, cmap='Greys')
plt.subplot(2,2,3)
plt.imshow(flippedLR, cmap='Blues')
plt.subplot(2,2,4)
plt.imshow(flippedUD, cmap='hsv')
```

```python
rotated = ndimage.rotate(img,45, reshape = False)
plt.imshow(rotated, cmap='Purples')
```

### Filter

```python
uniform_filtered = ndimage.uniform_filter(img,size=9)
plt.imshow(uniform_filtered)
```

```python
gaussian_filtered = ndimage.gaussian_filter(img,sigma=4)
plt.imshow(gaussian_filtered)
```

```python
median_filtered = ndimage.median_filter(img, 5)
plt.imshow(median_filtered)
```

```python
sobel_img =ndimage.sobel(img, axis=0)
plt.imshow(sobel_img)
```
